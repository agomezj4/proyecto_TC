```python
import sys
import os
import pandas as pd
```

```python
sys.path.insert(0, '../src')
from src.funciones import random_forest_selection, conditional_entropy_selection, intersect_top_features
```

```python
sys.path.insert(0, '../')
from src.parametros import rf_selection, ce_selection, features
```

```python
#load data
dataTC= pd.read_csv('../data/processed/dataTC_processed.csv')
dataTC
```

```python
ids = dataTC['id'].copy()
target= dataTC['y'].copy()
dataTC.drop(columns=['id'], inplace=True)
```

```python
random_forest_selection = random_forest_selection(dataTC, rf_selection)
conditional_entropy_selection = conditional_entropy_selection(dataTC, ce_selection)
intersect_features = intersect_top_features(random_forest_selection, conditional_entropy_selection, features)
```

```python
intersect_features
```

```python
dataTC = dataTC[list(intersect_features.intersection(dataTC.columns))]
dataTC
```

```python
dataTC_feature = pd.concat([ids, dataTC, target], axis=1)
```

**Guardar DataFrame en formato .csv**

```python
#train_transformed.to_csv('train_processed.csv', index=False)
dataTC_feature.to_csv(os.path.join('../data/feature', 'dataTC_feature.csv'), index=False)
```
